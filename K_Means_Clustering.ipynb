```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_excel('/content/iris_usl_ul.xlsx')
```

```python
dataset.head()
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(2,20):
  km=KMeans(init='k-means++',n_clusters=i,random_state=42)
  km.fit(dataset)
  wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot([i for i in range(2,20)],wcss,marker='o')
plt.xlabel('no of clusters')
plt.xticks([i for i in range(2,20)])
plt.ylabel('wcss')
plt.grid()
plt.show()
```
