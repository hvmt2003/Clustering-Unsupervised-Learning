```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_excel('/content/iris_usl_ul.xlsx')
```

```python
dataset.head()
```

```python
# check if the data is linearly seperable or not
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
# yes it is linearly seperable
```

```python
import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(20,20))
sc.dendrogram(sc.linkage(dataset,method='single',metric='euclidean'))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac=AgglomerativeClustering(n_clusters=2,linkage='single')
```

```python
# Remove the 'Predict' column if it exists from a previous failed run
if 'Predict' in dataset.columns:
    dataset = dataset.drop('Predict', axis=1)

# Perform clustering and add the predictions
dataset['Predict'] = ac.fit_predict(dataset)
display(dataset)
```

```python
sns.pairplot(data=dataset,hue='Predict')
plt.show()
```
