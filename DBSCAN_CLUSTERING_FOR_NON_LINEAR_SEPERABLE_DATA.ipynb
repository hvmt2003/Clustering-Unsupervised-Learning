```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
```

```python
x,y=make_moons(n_samples=550,noise=0.05)
```

```python
df=pd.DataFrame({'data1':x[:,0],'data2':x[:,1]})
```

```python
df.head()
```

```python
sns.scatterplot(x='data1',y='data2',data=df)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
db=DBSCAN(eps=0.2,min_samples=5)
df['Predict']=db.fit_predict(df)
```

```python
df.head()
```

```python
sns.scatterplot(x='data1',y='data2',data=df,hue='Predict')
plt.show()
```
